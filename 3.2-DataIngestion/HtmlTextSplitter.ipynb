```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<html>
    <body>
        <h1>Main Title</h1>
        <p>This is the introduction paragraph.</p>

        <h2>Subsection 1</h2>
        <p>Details about subsection 1.</p>

        <h2>Subsection 2</h2>
        <p>Details about subsection 2.</p>
    </body>
</html>
"""

# Headers to split on (tag, metadata key)
headers_to_split_on = [
    ("h1", "header_1"),
    ("h2", "header_2"),
]

text_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

docs = text_splitter.split_text(html_string)

# Output
for doc in docs:
    print("CONTENT:\n", doc.page_content)
    print("METADATA:\n", doc.metadata)
    print("-" * 50)
 
    
```
